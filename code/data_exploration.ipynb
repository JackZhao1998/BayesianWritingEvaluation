```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#import data
training_df = pd.read_csv('data/train.csv')
testing_df = pd.read_csv('data/test.csv')

training_df.columns
```

```python
training_df
```

```python
for i in range(2,len(training_df.columns)):
    term = training_df.columns[i]
    x = training_df[term]
    
    mean = np.mean(x)
    std = np.std(x)
    
    fit_x = np.linspace(1,5,500)
    fit_y = stats.norm.pdf(fit_x,mean,std)
    
    
    plt.figure()
    plt.hist(x,density = True,bins = 8)
    plt.plot(fit_x,fit_y)
    plt.title(term+' scores')
    
    plt.show()
```
